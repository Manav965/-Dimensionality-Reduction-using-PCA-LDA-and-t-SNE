```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
iris=load_iris()
iris.feature_names
iris.target
```

```python
x=iris.data
y=iris.target
y
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
scaled_data=scaler.transform(x)
scaled_data[0]
```

```python
pca=PCA(n_components=2)
pca.fit(scaled_data)
x_pca=pca.transform(scaled_data)
pca.explained_variance_ratio_
x_pca.shape
```

```python
pca=PCA(n_components=2)
pca.fit(x)
x_pc=pca.transform(x)
pca.explained_variance_ratio_
```

```python
x_pc.shape
```

```python
#The two prinicipal components 
pca1=x_pca[:,0]
pca2=x_pca[:,1]
```

```python
pca1=x_pc[:,0]
pca2=x_pc[:,1]
```

```python
#x_pca
sns.scatterplot(pca1,pca2,hue=y)
plt.xlabel("pca1")
plt.ylabel("pca2")
plt.title("Iris Data projected along direction with max variance")

```

```python
#TSNE portion
from sklearn.manifold import TSNE
x_t3eu=TSNE(n_components=3,metric='euclidean').fit_transform(x)
x_t2eu=TSNE(n_components=2,metric='euclidean').fit_transform(x)
x_t3ma=TSNE(n_components=3,metric='manhattan').fit_transform(x)
x_t2ma=TSNE(n_components=2,metric='manhattan').fit_transform(x)
```

```python
t1=x_t2eu[:,0]
t2=x_t2eu[:,1]

sns.scatterplot(t1,t2,hue=y)
plt.xlabel("tsne-1")
plt.ylabel("tsne-2")
plt.title("2D projection keeping euclidean distance as the metric parameter")
plt.show()
```

```python
tsne1=x_t3eu[:,0]
tsne2=x_t3eu[:,1]
tsne3=x_t3eu[:,2]
# Dataset
df=pd.DataFrame()
df['X']=tsne1
df['Y']=tsne2
df['Z']=tsne3
df['color']=y
 
# plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'],zdir='z', c=df['color'])

ax.set_xlim3d(-150,150)
ax.set_ylim3d(-150,150)
ax.set_zlim3d(-150,150)
ax.set_xlabel("tsne-1")
ax.set_ylabel("tsne-2")
ax.set_zlabel("tsne-3")

plt.title("3D projection keeping euclidean distance as the  metric parameter")
plt.show()
```
